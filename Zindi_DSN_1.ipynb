```python
def metric(y, pred):
    return roc_auc_score(y, pred, labels=[0,1])
```

```python

# kf = StratifiedKFold(n_splits=5,shuffle=True,random_state=seed)
# pred_test_full =0
# cv_score =[]
# i=1
# for train_index,test_index in kf.split(X,y):
#     print('{} of KFold {}'.format(i,kf.n_splits))
#     xtr,xvl = X.loc[train_index],X.loc[test_index]
#     ytr,yvl = y.loc[train_index],y.loc[test_index]
    
#     #model
#     lr = LogisticRegression(C=2)
#     lr.fit(xtr,ytr)
#     score = roc_auc_score(yvl,lr.predict(xvl))
#     print('ROC AUC score:',score)
#     cv_score.append(score)    
#     pred_test = lr.predict_proba(x_test)[:,1]
#     pred_test_full +=pred_test
#     i+=1


from catboost import CatBoostClassifier
from sklearn.metrics import log_loss, roc_auc_score


n_skf = 7
score_list = []
score=0
test_oofs = []

from sklearn.model_selection import StratifiedKFold
skf = StratifiedKFold(n_splits=7)

skf.get_n_splits(X, y)
print(skf)

i = 0
for train_index, test_index in skf.split(X, y):
    print("TRAIN:", train_index, "TEST:", test_index)
    X_train, X_test = X.loc[train_index], X.loc[test_index]
    y_train, y_test = y.loc[train_index], y.loc[test_index]
    
    
    model_cb = CatBoostClassifier(n_estimators = 4000,cat_features = categorical_indices, random_seed = 2020,
                                  learning_rate = 0.01, max_depth = 8,
                              eval_metric = "AUC", verbose = 100 , objective ="CrossEntropy",
                                  early_stopping_rounds = 200, use_best_model = True)
                                
    
    model_cb.fit(X_train, y_train, eval_set = (X_test, y_test), use_best_model = True, verbose = 100)
    
    
    p = model_cb.predict_proba(X_test)[:, 1]
    
    sc = metric(y_test, p)
    score_list.append(sc)
    score+= sc/n_skf
    
    i+= 1
    
    pred = model_cb.predict_proba(test[features])[:, 1]
    test_oofs.append(pred)
    print('Fold {} : {}'. format(i, sc))
    
print()
print()
print('Avg log :', score)    
```
